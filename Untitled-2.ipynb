````python
from pathlib import Path
import tifffile 
from cellpose import models, core 
from skimage.measure import regionprops
from skimage.measure import find_contours
from skimage import morphology
import numpy as np
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt


print('Using GPU', core.use_gpu())

def does_not_touch_image_border(prop, shape):
    """Return true if the roi does not touch the image border defined by the shape [ny,nx]"""  
    if len(prop.bbox) != 4:
        print('prop has not 4 elements')
    return (prop.bbox[0] > 0) and (prop.bbox[1] > 0) and (prop.bbox[2] < shape[1] and prop.bbox[3] < shape[0])    

def spatial_spread(prop, image):
    """Spread as the trace of the moment matrix in a region"""
    x,y = np.meshgrid(         
        np.arange(prop.bbox[1],prop.bbox[3]),
        np.arange(prop.bbox[0],prop.bbox[2]))    
    image = image[:,prop.bbox[0]:prop.bbox[2], prop.bbox[1]:prop.bbox[3]]
    L = []
    for weight in image:
        w = weight * prop.image
        if (w.max()-w.min()) > 0 :
            w = (w - w.min()) / (w.max() - w.min())
        else:
            return 0
        sw = np.sum(w)
        if sw < 1e-9:
            return 0.0
        sx = np.sum(w * x) / sw
        sy = np.sum(w * y) / sw
        sxx = np.sum(w * np.square(x-sx)) / sw
        syy = np.sum(w * np.square(y-sy)) / sw
        #sxy = np.sum(w * (x-sx) * (y-sy)) / sw
        L.append(np.sqrt(sxx + syy))
    return L


def measure_spread_in_image(model, img, channels):
    print('segment')
    mask = model.eval(
            img,
            channel_axis = 0, 
            channels = [channels['cells']+1, channels['nuclei']+1],
            diameter = 200, 
            flow_threshold = -1, 
            cellprob_threshold = 0.0, # default is 0
            min_size = 10 # filter out smaller cells
        )[0]

    print('regions')
    props = regionprops(mask)
    props = [x for x in props if does_not_touch_image_border(x, img.shape[1:3])]

    print('spread')
    se = morphology.disk(30)

    signal = np.stack([morphology.white_tophat(ch, se) for ch in img])

    spread = np.array([spatial_spread(p, signal) for p in props])        

    tmp =  pd.DataFrame({
        
        'Cell': np.arange(len(props)),
        'Area': [x.area for x in props]
        })
    
    for k, c in enumerate(channels):
        tmp['spread [' + c + ']'] = spread[:,k] 

    return tmp, props    

def measure_spread_in_cells(filename, channels):
    imglist = tifffile.imread(filename)
    model = models.Cellpose(gpu=core.use_gpu(), model_type='cyto2')
    df = []
    rois = [] * len(imglist)
    for idx, img in enumerate(imglist):
        tmp, props = measure_spread_in_image(model, img, channels)
        tmp['Filename'] = [filename] * len(props)
        tmp['Image'] = [ idx ] * len(props)
        df.append(tmp)
        rois[idx] = props
        
    return pd.concat(df), rois

def visualize(img, rois, df):
    from matplotlib.patches import Ellipse        
    plt.imshow(img[0])
    for r in rois:
        c = find_contours(np.pad(r.image,1) > 0, 0.5)
        plt.plot(c[0][:,1] + r.bbox[0], c[0][:,0] + r.bbox[1])
        #ellipse = Ellipse((p.centroid[0], p.centroid[1]), width = df
    plt.axis('off')        
````

````python
from cellpose import models, core 
from skimage.measure import regionprops
from skimage.measure import find_contours
import numpy as np
import matplotlib.pyplot as plt

def segment_cells(img,type=0):
    """
    Parameters:
    img : image with s
    type : 0 segment nuclei and watershed, 1: cellpose
    """
    print('segment')
    from scipy.ndimage import distance_transform_edt, median_filter,label
    from skimage.segmentation import watershed
    if type == 0:
        model = models.Cellpose(gpu=core.use_gpu(), model_type='nuclei')    
        nlabels = model.eval(
                img,
                channel_axis = 0, 
                channels = [1,0],
                diameter = 250, 
                flow_threshold = 0.4, 
                cellprob_threshold = 0.1, # default is 0
                min_size = 100000 # filter out smaller cells
            )[0]                                        
        dist = distance_transform_edt(nlabels > 0)
        mask = median_filter(np.amax(img,0), 11)
        mask = mask > mask[mask < np.quantile(mask, 0.1)].mean()
        clabels = watershed(-dist, nlabels, mask=mask)            
    else :
        model = models.Cellpose(gpu=core.use_gpu(), model_type='cyto2')    
        clabels = model.eval(
                img,
                channel_axis = 0,
                channels = [2,4],
                diameter = 250, 
                flow_threshold = 0.4, 
                cellprob_threshold = -1, # default is 0
                min_size = 100000 # filter out smaller cells
            )[0]             
    props = regionprops(clabels)                             
    props = [x for x in props if does_not_touch_image_border(x, img.shape[1:3])]
    #props = [x for x in props if x.area > 10000]
    return props

def visualize(img, rois, s=None):
    from matplotlib.patches import Circle
    #plt.imshow(img[[2,1,0]].swapaxes(0,-1))
    plt.imshow(img[0],cmap='gray')
    for k,r in enumerate(rois):
        c = find_contours(np.pad(r.image,1) > 0, 0.5)
        plt.plot(c[0][:,1] + r.bbox[1], c[0][:,0] + r.bbox[0]) 
        if s is not None:
            c = plt.Circle((r.centroid[0], r.centroid[1]), s[k], fill=False)            
            plt.gca().add_patch(c)
    plt.axis('off')


roi = segment_cells(img,0)
visualize(img, roi)
````

````python
from scipy.ndimage import minimum_filter, maximum_filter, gaussian_filter

def spatial_spread(prop, image):
    """Spread as the trace of the moment matrix in a region for all channels
    Result
    ------
    list of tuple (centroid x, centroid y, sxx, sxy, syy) for each channel
    """
    y,x = np.meshgrid(         
        np.arange(prop.bbox[0],prop.bbox[2]),
        np.arange(prop.bbox[1],prop.bbox[3]),indexing='ij')    
    image = image[:,prop.bbox[0]:prop.bbox[2], prop.bbox[1]:prop.bbox[3]]
    S = []        
    for weight in image:
        w = weight *(weight>(weight.mean() + weight.std())) * prop.image        
        if (w.max()-w.min()) > 0 :
            w = (w - w.min()) / (w.max() - w.min())
        else:
            w = prop.image
        sw = np.sum(w)
        if sw < 1e-9:
            return 0.0
        sx = np.sum(w * x) / sw
        sy = np.sum(w * y) / sw
        sxx = np.sum(w * np.square(x-sx)) / sw
        syy = np.sum(w * np.square(y-sy)) / sw
        sxy = np.sum(w * (x-sx) * (y-sy)) / sw
        S.append((sx,sy,sxx,sxy,syy))              
    return S

def spatial_spread_mask(mask, image):
    """Spread as the trace of the moment matrix in a region for all channels
    Result
    ------
    list of tuple (centroid x, centroid y, sxx, sxy, syy) for each channel
    """
    y,x = np.meshgrid(np.arange(mask.shape[1]), np.arange(mask.shape[0]),indexing='ij')    
    w = image *(image>(image.mean() + image.std())) * mask
    if (w.max()-w.min()) > 0 :
        w = (w - w.min()) / (w.max() - w.min())
    else:
        w = prop.image
    sw = np.sum(w)
    if sw < 1e-9:
        return 0.0
    sx = np.sum(w * x) / sw
    sy = np.sum(w * y) / sw
    sxx = np.sum(w * np.square(x-sx)) / sw
    syy = np.sum(w * np.square(y-sy)) / sw
    sxy = np.sum(w * (x-sx) * (y-sy)) / sw
    return sx,sy,sxx,sxy,syy


# prefilter the image by a Gaussian filter
signal = np.stack([gaussian_filter(ch, 10) for ch in img])

# compute the spread in all regions and all channels 
S = np.array([spatial_spread(x, signal) for x in roi ]) # S: ROI, Channel, Moments
````
